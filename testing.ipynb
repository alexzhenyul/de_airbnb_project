```python
import pandas as pd
```

```python
mel_listing_df = pd.read_csv("data/listings.csv")
```

```python
mel_listing_df.info()
```

```python
# Modify dataframe into fact & dimension table

# dimension table for airbnb listings
dim_listings_columns = ['id', 'listing_url', 'host_id', 'name', 'description', 'neighborhood_overview', 'picture_url', 'latitude', 'longitude', 'property_type', 'room_type', 'accommodates', 'bathrooms_text', 'beds', 'amenities']
dim_listings = mel_listing_df[dim_listings_columns]
dim_listings = dim_listings.rename(columns={'id': 'listing_id', 'name': 'listing_name', 'description': 'listing_description'}).sort_values(by=['listing_id'], ascending=True)
dim_listings.info()

## bathrooms field is evolved from number to textual description (Example: 2 bathrooms/1 Shared bathroom)
```

```python
# dimension table for airbnb hosts 
dim_hosts = mel_listing_df.iloc[:, 9:27]
dim_hosts
dim_hosts.info()
```

```python
# Dimension table for review stats
dim_review_stats = mel_listing_df.iloc[:, [0] + list(range(56, len(mel_listing_df.columns)))]
dim_review_stats = dim_review_stats.rename(columns={'id': 'listing_id'}).sort_values(by=['listing_id'], ascending=True)
dim_review_stats
dim_review_stats.info()
```

```python
mel_review_df = pd.read_csv("data/reviews.csv")
mel_review_df.info()
#mel_review_df
```

```python
# dimension table for airbnb reviewers
dim_reviewers_columns = ['reviewer_id', 'reviewer_name', 'id']
dim_reviewers = mel_review_df[dim_reviewers_columns]
dim_reviewers = dim_reviewers.rename(columns={'id': 'comment_id'}).sort_values(by=['comment_id'], ascending=True)
dim_reviewers.info()
#dim_reviewers
```

```python
# Fact table for airbnb reviews
dim_reviews_columns = ['id', 'reviewer_id', 'listing_id', 'date','comments']
dim_reviews = mel_review_df[dim_reviews_columns]
dim_reviews = dim_reviews.rename(columns={'id': 'comment_id', 'date': 'comment_date'}).sort_values(by=['comment_id'], ascending=True)
dim_reviews.info()
```

```python
mel_calendar_df = pd.read_csv("data/calendar.csv")
mel_calendar_df.info()
```

```python
mel_calendar_df['date'] = pd.to_datetime(mel_calendar_df['date'])
mel_calendar_df.head()

# Data enrichment for datetime in various level
mel_calendar_df['year'] = mel_calendar_df['date'].dt.year
mel_calendar_df['month'] = mel_calendar_df['date'].dt.month
mel_calendar_df['day'] = mel_calendar_df['date'].dt.day
mel_calendar_df['quarter'] = mel_calendar_df['date'].dt.quarter
mel_calendar_df['weekday'] = mel_calendar_df['date'].dt.weekday
mel_calendar_df['week'] = mel_calendar_df['date'].dt.week
mel_calendar_df
```

```python
# Dimension table for datetime
dim_datetime_columns = ['date', 'year', 'month', 'day','quarter', 'weekday', 'week']
dim_datetime = mel_calendar_df[dim_datetime_columns]
dim_datetime['datetime_id'] = dim_datetime.index
dim_datetime = dim_datetime[['datetime_id', 'date', 'year', 'month','day','quarter', 'weekday', 'week']]
dim_datetime.info()
dim_datetime.head()
```

```python
result = mel_calendar_df[[col for col in mel_calendar_df.columns if col not in dim_datetime_columns]]
result
```
